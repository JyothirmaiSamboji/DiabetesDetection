```python
import numpy as np
import pandas as pd
df=pd.read_csv("diabities.csv")
del df['Unnamed: 0']
df.head(10)
```

```python
#to find if there r any null contraints for columns so that we can remove them
df.info()
```

```python
#to find if there r any missing elements or NaN
df.isnull().sum()
```

```python
df.head(10)
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=25)
```

```python
#building model using logistic regression
from sklearn.linear_model import LogisticRegression
LRmodel=LogisticRegression(solver='lbfgs',max_iter=1000).fit(x_train,y_train)
y_predict=LRmodel.predict(x_test)
#for model evaluation
from sklearn.metrics import accuracy_score
acc=round(accuracy_score(y_predict,y_test),2)*100
print("Accuracy of detecting persons with diabetics through log reg:  ",acc)
```

```python
#building model using random forest
from sklearn.ensemble import RandomForestClassifier
RFmodel=RandomForestClassifier(criterion='entropy',n_estimators=9,random_state=1)
RFmodel.fit(x_train,y_train)
y_predict1=RFmodel.predict(x_test)
acc=round(accuracy_score(y_predict1,y_test),2)*100
print("Accuracy of detecting persons with diabetics througAh random forest:  ",acc)
```

```python
#building model using k nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
KNNmodel=KNeighborsClassifier(n_neighbors=3)
KNNmodel.fit(x_train,y_train)
y_predict2=KNNmodel.predict(x_test)
acc=round(accuracy_score(y_predict2,y_test),2)*100
print("Accuracy of detecting persons with diabetics through k nearest neighbors :  ",acc)
```
